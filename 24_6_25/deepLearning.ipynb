```python
import matplotlib.pyplot as plt
```

## ReLU

### Hard Coded

```python
x = [i for i in range(-10,10)]
```

```python
plt.plot(x)
```

```python
def rectifiedLinearUnit(x):
  return max(0.0,x)

```

```python
output = [rectifiedLinearUnit(i) for i in x]
```

```python
plt.plot(x,output)
```

### Torch

```python
import torch
from torch.nn import ReLU
```

```python
tensor_X = torch.range(start=-10,end=10,step=1)
relu_object = ReLU()
torch_output = relu_object(tensor_X)
torch_output
```

### TensorFlow

```python
import tensorflow as tf
from tensorflow.nn import relu

tf_tensor = tf.range(start=-10, limit=10)
tf_output = relu(tf_tensor).numpy()
tf_output
```

## TanH


### Hard Coded

```python
from math import exp
x = [i for i in range(-10,10)]

def hyperbolicTangent(x):
  return (exp(x) - exp(-x))/(exp(x) + exp(-x))

output = [hyperbolicTangent(i) for i in x]
```

```python
plt.plot(x,output)
```

### Torch

```python
from torch.nn import Tanh
```

```python
tensor_X = torch.range(start=-10,end=10,step=1)
tanh_object = Tanh()
torch_output = tanh_object(tensor_X)
torch_output
```

### TensorFlow

```python
from tensorflow.nn import tanh

tf_tensor = tf.range(start=-10.0, limit=10.0)
tf_output = tanh(tf_tensor).numpy()
tf_output
```

# SoftMax Activation

### Hard Coded

```python
import numpy as np
np.random.seed(7)

output_layer_vector = np.random.randn(1,9)
output_layer_vector
```

```python
def softmax(x):
  return np.exp(x) / np.exp(x).sum()
```

```python
result = softmax(output_layer_vector)
result
```

```python
np.argmax(result)
```

### Torch

```python
from torch.nn import Softmax
torch.manual_seed(7)
x = torch.randn([1,5])
softmax = Softmax()
output = softmax(x)
print(f'Output Vectors: {x}')
print(f'Softmax Vectors: {output}')
```

### TensorFlow

```python
from tensorflow.nn import softmax

tf.random.set_seed(7)
x = tf.random.normal([1,5], dtype=tf.float64)
output = softmax(x)
print(f'Output Vectors: {x}')
print(f'Softmax Vectors: {output}')
```
